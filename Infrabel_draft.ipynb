```python
import pandas as pd
import numpy as np
import csv
import json
import math
```

```python
dfS = pd.read_csv("stations.csv", sep = ';')
dfS = dfS.where(dfS['Classification en anglais']=='Gare').dropna()
dfL = pd.read_csv("geosporen.csv", sep = ';')

LID = dfL['Identifiant de la voie'].unique()
SID = dfS['Nom symbolique'].unique()
nL, nS = len(LID), len(SID)
```

```python
A = np.zeros((nL, nS))

for s in dfS.values:
    p = [float(x) for x in s[0].split(',')]
    Lanes = set()
    for l in dfL.values:
        if isinstance(l[1], str):
            gs = json.loads(str(l[1]))
            for p2 in gs['coordinates']:
                p2 = [p2[1], p2[0]]
                if not isinstance(p2[1], list) and math.dist(p, p2) < 10e-3: Lanes.add(l[2])
    print(s[4], Lanes)
```
